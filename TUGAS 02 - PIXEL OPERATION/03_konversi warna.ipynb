# Konversi Warna

Konversi warna, dalam konteks pemrosesan gambar, adalah transformasi skema warna gambar dari satu ruang warna ke ruang warna lainnya, misalnya, dari RGB (Red, Green, Blue) ke skala abu-abu.

### Konversi RGB ke GRAYSCALE (Skala Abu-abu)

Grayscale adalah kisaran warna abu-abu tanpa warna yang jelas. Warna yang paling gelap adalah hitam, yaitu ketiadaan sama sekali cahaya yang ditransmisikan atau dipantulkan. Warna yang paling terang adalah putih, yaitu transmisi atau pantulan total cahaya pada semua panjang gelombang yang terlihat.

Dalam pencitraan digital, skala abu-abu diturunkan dari model warna RGB. Model warna RGB adalah model warna aditif, di mana cahaya merah, hijau, dan biru ditambahkan bersama dalam berbagai cara untuk mereproduksi beragam warna.

Konversi dari RGB ke skala abu-abu dilakukan dengan menggunakan rumus:
`grayscale = 0.2989 * R + 0.5870 * G + 0.1140 * B`

Formula ini adalah metode standar untuk mengonversi RGB ke skala abu-abu yang memperhitungkan berbagai cara berbeda yang digunakan manusia untuk melihat warna. Koefisien (0,2989, 0,5870, dan 0,1140) mewakili persepsi intensitas terukur dari penglihatan trichromat khas manusia, yang paling peka terhadap warna hijau dan paling tidak peka terhadap warna biru.

Untuk mengonversi seluruh gambar ke skala abu-abu, rumus ini diterapkan ke setiap piksel dalam gambar. Gambar skala abu-abu yang dihasilkan adalah larik 2D nilai skala abu-abu.

```python
def rgb_to_grayscale(red, green, blue):
    return 0.2989 * red + 0.5870 * green + 0.1140 * blue

def image_rgb_to_grayscale(image):
    # representasi gambar ke array 3D 
    # 3d array = [height][width][RGB]
    height, width, _ = image.shape
    
    # List grayscale_image 2D baru dibuat dengan tinggi dan lebar yang sama dengan gambar input.
    # ini akan menjadi ukuran output gambar
    grayscale_image = [[0 for _ in range(width)] for _ in range(height)]
    
    # Looping setiap piksel dalam gambar input. Untuk setiap piksel, fungsi ini menguraikan nilai RGB ke dalam variabel red, green, dan blue.
    for i in range(height):
        for j in range(width):
            red, green, blue = image[i][j]
            grayscale_image[i][j] = rgb_to_grayscale(red, green, blue)
    return grayscale_image
```

```python
import matplotlib.pyplot as plt
from cv2 import imread, cvtColor, COLOR_BGR2RGB

# Load gambar
image_path = '../FILE/aang.jpg'
image = imread(image_path)
grayscale_img = image_rgb_to_grayscale(image)

# bandingkan dengan original image
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(cvtColor(image, COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# tampikan gambar grayscale
plt.subplot(1, 2, 2)
plt.imshow(grayscale_img, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```
