# Kecerahan gambar

Kecerahan gambar adalah ukuran cahaya yang jatuh pada gambar. Pada gambar digital, kecerahan dapat disesuaikan dengan mengubah nilai piksel dalam gambar.

Tiap piksel dalam gambar memiliki nilai kecerahan, yang bisa direpresentasikan dalam ruang warna RGB (Red, Green, Blue). Kecerahan piksel adalah nilai skalar yang merepresentasikan intensitas cahaya. 

Untuk meningkatkan kecerahan gambar, kita dapat menambahkan nilai konstan ke nilai RGB semua piksel dalam gambar. Konstanta ini disebut "faktor kecerahan". Jika faktor kecerahannya positif, gambar akan lebih cerah. Jika negatif, gambar akan lebih gelap.

Contohnya, jika kita memiliki piksel dengan nilai RGB (100, 100, 100) dan menambahkan faktor kecerahan 50, maka nilai RGB baru piksel tersebut adalah (150, 150, 150), yaitu warna yang lebih cerah.

Namun demikian, penting untuk diperhatikan, bahwa nilai RGB harus berada dalam kisaran 0-255. Jika suatu operasi membuat nilai RGB kurang dari 0 atau lebih dari 255, kita harus memotongnya menjadi 0 atau 255.

Ini adalah metode yang sederhana dan mudah, tetapi mungkin tidak selalu memberikan hasil terbaik, karena tidak memperhitungkan fakta bahwa persepsi manusia tentang kecerahan adalah nonlinier dan bervariasi, tergantung pada warnanya.


```python
def modify_brightness(image, brightness):
    # Looping tiap pixel yang ada di gambar
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            # Mengambil nilai RGB dari tiap pixel
            r, g, b = image[i, j]

            # Ubah kecerahan dengan menambahkan nilai kecerahan ke setiap channel
            r += brightness
            g += brightness
            b += brightness

            # Pastikan nilainya berada dalam kisaran yang valid antara 0-255
            r = max(0, min(255, r))
            g = max(0, min(255, g))
            b = max(0, min(255, b))

            # Memperbarui nilai piksel
            image[i, j] = [r, g, b]

    return image

```

```python
import matplotlib.pyplot as plt
from cv2 import imread, cvtColor, COLOR_BGR2RGB

image_path = "../FILE/aang.jpg"
image_original = imread(image_path)

plt.imshow(cvtColor(image_original, COLOR_BGR2RGB))
plt.title('Gambar original')
plt.axis('off')
plt.show()
```

```python
# modifikasi kecerahan gambar 
image_path = "../FILE/aang.jpg"
image_original = imread(image_path)
modified_image = modify_brightness(image_original, 100)
modified_image_rgb = cvtColor(modified_image, COLOR_BGR2RGB)


plt.imshow(modified_image_rgb)
plt.title('Gambar dengan Brightness +100')
plt.axis('off')
plt.show()
```
