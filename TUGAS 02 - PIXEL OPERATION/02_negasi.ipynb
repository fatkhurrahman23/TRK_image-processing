# Negasi gambar

Negasi gambar, juga dikenal sebagai inversi gambar atau negatif, adalah proses yang membalikkan warna pada gambar. 

Dalam ruang warna RGB (Red, Green, Blue), nilai maksimum untuk setiap saluran warna adalah 255. Untuk meniadakan suatu gambar, kita mengurangi nilai saat ini dari setiap saluran warna dari 255. Operasi ini membalikkan saluran warna. Sebagai contoh, piksel dengan nilai RGB (100, 100, 100) akan menjadi (155, 155, 155) setelah dinegasikan.

Rumus untuk negasi adalah: <br>
`nilai_piksel_negasi = 255 - nilai_piksel_asli`

Operasi ini dilakukan pada setiap piksel dalam gambar, secara efektif menciptakan negatif dari gambar aslinya. 

Pada gambar yang dihasilkan, area gelap menjadi terang dan area terang menjadi gelap. Warna juga dibalik, misalnya, merah menjadi cyan, hijau menjadi magenta, dan biru menjadi kuning.

```python
import numpy as np

def negate_image(image):
    # Menambil dimensi gambar
    height, width = image.shape[:2]
    
    # Membuat array kosong untuk menyimpan gambar yang dinegasikan
    negated_image = np.zeros((height, width, 3), dtype=np.uint8)
    
    # Looping setiap pixel yang ada pada gambar
    for i in range(height):
        for j in range(width):
            # ambil nilai RGB dari piksel saat ini
            r, g, b = image[i, j]
            
            # Negate the RGB values
            negated_r = 255 - r
            negated_g = 255 - g
            negated_b = 255 - b
            
            # Menetapkan nilai RGB yang dinegasikan dalam gambar baru
            negated_image[i, j] = (negated_r, negated_g, negated_b)
    
    return negated_image
```

```python
import matplotlib.pyplot as plt
from cv2 import imread, cvtColor, COLOR_BGR2RGB

# Load the image
image_path = '../FILE/aang.jpg'
image = imread(image_path)
negate_image = negate_image(image)
negated_image = cvtColor(negate_image, COLOR_BGR2RGB)

# bandingkan dengan original image
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(cvtColor(image, COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# tampikan gambar grayscale
plt.subplot(1, 2, 2)
plt.imshow(negated_image)
plt.title('Negative Image')
plt.axis('off')
plt.show()


```
