```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
# load file image
img = cv2.imread("/content/drives/MyDrive/FILE COLAB/IMAGE PROCESSING/aang.jpg")
cv2_imshow(img)
```

```python
grayscale_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(grayscale_img)
```

```python
# menggunakan binary threshold dari gambar grayscale
_, bw = cv2.threshold(grayscale_img, 127, 255, cv2.THRESH_BINARY)
cv2_imshow(bw)
```
